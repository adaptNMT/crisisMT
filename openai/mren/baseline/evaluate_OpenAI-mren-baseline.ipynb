```
from openai import OpenAI
client = OpenAI()
```

```
# gpt-3.5-turbo (oringally - now swapped in my custom model)
# open the test file for reading, src-test.txt
# open the pred file for writing, pred.txt
src= open('src-test.txt', 'r')
pred = open('pred.txt', 'w')

Lines = src.readlines()
count = 0
# Strips the newline character
for line in Lines:
    count += 1
    srcl=line.strip()
    print(srcl)

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        messages=[{"role": "system", "content": "Translate Marathi to English."},
                  {"role": "user", "content": srcl}],
        temperature=0.5
    )

    translated_text = completion.choices[0].message.content
    print(translated_text) # display translations in real time
    pred.writelines(translated_text + '\n')

src.close()
pred.close()
!ls
```
